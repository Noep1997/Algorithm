```python
import pandas as pd
import numpy as np
from sklearn.metrics import log_loss
from sklearn.feature_extraction.text import CountVectorizer
from IPython.display import display
```

```python
#load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

#separate the response 
train_response = train[['toxic','severe_toxic','obscene','threat','insult','identity_hate']]
train_input = train[['comment_text']]

#countvectorizer
cv = CountVectorizer()
x_train = cv.fit_transform(train.comment_text.values)
x_test = cv.transform(test.comment_text.values)
```

```python
#display the head of the training data
train.head()
```

```python
from sklearn.model_selection import train_test_split

# splitting the training data into train & test so we have some values to check against and gage the rate of success
x_train2, x_test2, y_train2, y_test2 = train_test_split(train_input, train_response, test_size=0.20)

#countvectorizer for these new subdatasets -- for the algorithm fitting formats to work
x_train2 = cv.fit_transform(x_train2.comment_text.values)
x_test2 = cv.transform(x_test2.comment_text.values)
```

```python
# we decided upon a gradient boosting classifier for its similarity to random forest, yet strong performance
from sklearn.ensemble import GradientBoostingClassifier

# output var
train_predict = pd.DataFrame(columns = train_response.columns)

#loop through the response and fit a classifier for each one
for col in train_response.columns:
    # fit for this column
    GBC = GradientBoostingClassifier()
    GBC.fit(x_train2, y_train2[col].values)
    train_predict[col] = GBC.predict_proba(x_test2)[:,1]
    
# display our prediction
display(train_predict)
```

```python
from sklearn.metrics import log_loss

# testing for log loss accuracy of the training subset
ll_score = log_loss(y_test2, train_predict)
display(ll_score) # 0.325 not bad, let's compare with the random forst
```

```python
# baseline random forest code for log loss comparison on modified input set
from sklearn.ensemble import RandomForestClassifier

train_predict2 = pd.DataFrame(columns = train_response.columns)

#loop through the response and fit a classifier for each one
for col in train_response.columns:
    rf = RandomForestClassifier()
    rf.fit(x_train2, y_train2[col].values)
    train_predict2[col] = rf.predict_proba(x_test2)[:,1]
    
# error
ll_score2 = log_loss(y_test2, train_predict2)
display(ll_score2) # as we can see, our fitting is a lot better --> mission accomplished
```

Clearly, 0.325 is a lot better than 0.980 so mission accomplished, we beat the baseline random forest. While this is not directly <0.199, we infer since the random forest has 0.199 on the full set that this will have much better than 0.199 log loss error on the full set (which we can't test remotely). We now run this better fit algorithm on our actual test sample, refitting with the whole dataset for even better accuracy.

```python
# running our better-fit algorithm on the actual test set

# real output var
test_predict = pd.DataFrame(columns = train_response.columns)

#loop through the response and fit a classifier for each one
for col in train_response.columns:
    # fit for this column
    GBC = GradientBoostingClassifier()
    GBC.fit(x_train, train_response[col].values)
    test_predict[col] = GBC.predict_proba(x_test)[:,1]
    
# display our prediction
display(train_predict)
```
